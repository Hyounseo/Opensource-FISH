```python
!pip install easyocr pandas tqdm

import os
import easyocr
import pandas as pd
from tqdm import tqdm

# 1. 로컬 EasyOCR 리더기 초기화 (GPU 사용)
reader = easyocr.Reader(['ko', 'en'], gpu=True)

# 2. 구글 드라이브 실제 폴더 경로 정확히 지정
correct_folder = '/content/drive/MyDrive/AI_OpenSourceProgramming_Correct'
phishing_folder = '/content/drive/MyDrive/AI_Phishing_Project'

print(f"🎯 지정된 정상 경로: {correct_folder}")
print(f"🎯 지정된 피싱 경로: {phishing_folder}\n")

# 데이터를 담을 빈 리스트 생성
data_list = []

# --- [3. 정상 데이터 추출 (Label: 0)] ---
print("--- [1/2] 정상 데이터 추출 시작 ---")
if os.path.exists(correct_folder):
    for filename in tqdm(os.listdir(correct_folder)):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
            image_path = os.path.join(correct_folder, filename)
            try:
                result = reader.readtext(image_path, detail=0)
                extracted_text = " ".join(result)
                data_list.append({'text': extracted_text, 'label': 0})
            except Exception as e:
                pass
else:
    print("❌ 정상 폴더 경로를 찾을 수 없습니다. 구글 드라이브를 확인해 주세요.")

# --- [4. 피싱 데이터 추출 (Label: 1)] ---
print("\n--- [2/2] 피싱 데이터 추출 시작 ---")
if os.path.exists(phishing_folder):
    for filename in tqdm(os.listdir(phishing_folder)):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
            image_path = os.path.join(phishing_folder, filename)
            try:
                result = reader.readtext(image_path, detail=0)
                extracted_text = " ".join(result)
                # 피싱 데이터이므로 label을 1로 지정하여 추가
                data_list.append({'text': extracted_text, 'label': 1})
            except Exception as e:
                print(f"피싱 파일 에러 ({filename}): {e}")
else:
    print("❌ 피싱 폴더 경로를 찾지 못했습니다. 폴더명이 정확한지 다시 확인해 주세요.")


# --- [5. 판다스 DataFrame 변환 및 분포 확인] ---
if data_list:
    df = pd.DataFrame(data_list)
    print("\n=== ✨ 자연어 처리(NLP) 데이터셋 구축 완료! ===")
    print(f"총 수집된 데이터 개수: {len(df)}개")
    print("\n[라벨별 데이터 분포 (0: 정상, 1: 피싱)]")
    print(df['label'].value_counts())

    # 데이터가 잘 들어갔는지 샘플 5개 슬쩍 보기
    print("\n[데이터 샘플 미리보기]")
    print(df.sample(min(5, len(df))))
else:
    print("\n❌ 추출된 데이터가 없습니다. 경로를 다시 점검해야 합니다.")
```
