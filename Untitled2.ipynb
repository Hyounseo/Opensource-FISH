```python
# 1. 코랩 내부에 업로드한 dataset.zip.zip 이름 그대로 압축 해제
!unzip -q /content/dataset.zip.zip -d /content/

import os
import easyocr
import pandas as pd
from tqdm import tqdm

# 2. 로컬 EasyOCR 리더기 초기화 (GPU 사용)
reader = easyocr.Reader(['ko', 'en'], gpu=True)

# 3. 로컬 폴더 경로 지정
correct_folder = '/content/correct'
phishing_folder = '/content/phishing'

data_list = []

# --- 정상 데이터 추출 (Label: 0) ---
print("--- [1/2] 정상 데이터 추출 시작 ---")
if os.path.exists(correct_folder):
    for filename in tqdm(os.listdir(correct_folder)):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
            image_path = os.path.join(correct_folder, filename)
            try:
                result = reader.readtext(image_path, detail=0)
                data_list.append({'text': " ".join(result), 'label': 0})
            except Exception as e:
                pass
else:
    print("❌ correct 폴더를 찾을 수 없습니다. 압축파일 내부 폴더명을 확인해 주세요.")

# --- 피싱 데이터 추출 (Label: 1) ---
print("\n--- [2/2] 피싱 데이터 추출 시작 ---")
if os.path.exists(phishing_folder):
    for filename in tqdm(os.listdir(phishing_folder)):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
            image_path = os.path.join(phishing_folder, filename)
            try:
                result = reader.readtext(image_path, detail=0)
                data_list.append({'text': " ".join(result), 'label': 1})
            except Exception as e:
                pass
else:
    print("❌ phishing 폴더를 찾을 수 없습니다. 압축파일 내부 폴더명을 확인해 주세요.")

# --- 데이터셋 최종 생성 ---
if data_list:
    df = pd.DataFrame(data_list)
    print("\n=== ✨ 자연어 처리(NLP) 데이터셋 구축 완료! ===")
    print(df['label'].value_counts())
else:
    print("\n❌ 데이터를 가져오지 못했습니다. 코랩 왼쪽 파일 창에 파일이 올바르게 올라갔는지 확인해 주세요.")
```

```python
# 현재 메모리에 있는 df를 인덱스 제외하고 깔끔하게 CSV로 저장
df.to_csv('/content/phishing_dataset.csv', index=False, encoding='utf-8-sig')
print("✅ 라벨링된 텍스트 데이터가 'phishing_dataset.csv' 파일로 완벽하게 저장되었습니다!")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score

# 1. 생성된 CSV 데이터셋 로드
csv_path = '/content/phishing_dataset.csv'
dataset = pd.read_csv(csv_path)

# 텍스트 데이터 중 빈 값이 있다면 빈 문자열('')로 대체 (결측치 예방)
dataset['text'] = dataset['text'].fillna('')

print(f"📊 CSV 기반 학습 데이터셋 로드 완료! (총 샘플 수: {len(dataset)}개)")
print(dataset['label'].value_counts())
print("-" * 50)

# 2. 학습 데이터(80%)와 검증용 테스트 데이터(20%)로 분리
# stratify=dataset['label']을 주어 두 그룹의 정상/피싱 비율이 골고루 섞이게 합니다.
X_train_raw, X_test_raw, y_train, y_test = train_test_split(
    dataset['text'], dataset['label'], test_size=0.2, random_state=42, stratify=dataset['label']
)

# 3. TF-IDF를 이용해 자연어 텍스트를 숫자 벡터(행렬)로 변환
# 중요 단어 상위 1000개만 뽑아서 컴퓨터가 이해할 수 있는 숫자로 매핑합니다.
tfidf = TfidfVectorizer(max_features=1000)
X_train = tfidf.fit_transform(X_train_raw)
X_test = tfidf.transform(X_test_raw)

# 4. 외부 AI API 없이 '스킷런(Scikit-learn)' 라이브러리의 로컬 머신러닝(로지스틱 회귀) 알고리즘 학습
nlp_model = LogisticRegression()
nlp_model.fit(X_train, y_train)

# 5. 성능 검증 및 최종 리포트 출력
y_pred = nlp_model.predict(X_test)
print(f"🎯 [최종 결과] 피싱 사이트 판별 정확도(Accuracy): {accuracy_score(y_test, y_pred) * 100:.2f}%")
print("\n--- 📝 상세 평가지표 ---")
print(classification_report(y_test, y_pred))
```
