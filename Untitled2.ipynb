```python
!pip install easyocr pandas tqdm

import os
import easyocr
import pandas as pd
from tqdm import tqdm

# 로컬 EasyOCR 리더기 초기화
reader = easyocr.Reader(['ko', 'en'], gpu=True)

# --- [자동 경로 탐색 로직] ---
base_path = '/content/drive/MyDrive/'
all_folders = os.listdir(base_path)

# 내 드라이브에서 해당 폴더명과 유사한 폴더 자동 매칭
matched_folder = None
for f in all_folders:
    if 'AI_OpenSourceProgramming' in f and 'Correct' in f: # 글자가 포함되어 있다면
        matched_folder = os.path.join(base_path, f)
        break

if matched_folder is None:
    # 만약 그래도 못 찾으면 수동으로 썼던 경로를 강제 지정
    correct_folder = '/content/drive/MyDrive/AI_OpenSourceProgramming_Correct'
else:
    correct_folder = matched_folder
    print(f"✅ 코랩이 실제 인식한 폴더 경로: {correct_folder}")
# ----------------------------

data_list = []

print("--- 정상 데이터 추출 시작 ---")
try:
    for filename in tqdm(os.listdir(correct_folder)):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
            image_path = os.path.join(correct_folder, filename)
            try:
                result = reader.readtext(image_path, detail=0)
                extracted_text = " ".join(result)
                data_list.append({'text': extracted_text, 'label': 0})
            except Exception as e:
                print(f"파일 에러 ({filename}): {e}")
except FileNotFoundError:
    print("\n❌ [비상] 코랩 시스템이 아직 구글 드라이브 폴더를 동기화하지 못했습니다.")
    print("💡 해결법: 코랩 왼쪽 메뉴에서 [런타임] -> [런타임 다시 시작]을 누른 후, 1단계(드라이브 마운트)부터 딱 한 번만 다시 실행해 주세요!")
```
