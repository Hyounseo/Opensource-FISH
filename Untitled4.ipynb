```python
import os
import re
import pandas as pd
import numpy as np

# [2단계-A] 교수님 요구사항 충족을 위한 편집 거리(Levenshtein Distance) 직접 구현
def get_levenshtein_distance(s1, s2):
    if len(s1) < len(s2):
        return get_levenshtein_distance(s2, s1)
    if len(s2) == 0:
        return len(s1)

    previous_row = range(len(s2) + 1)
    for i, c1 in enumerate(s1):
        current_row = [i + 1]
        for j, c2 in enumerate(s2):
            insertions = previous_row[j + 1] + 1
            deletions = current_row[j] + 1
            substitutions = previous_row[j] + (c1 != c2)
            current_row.append(min(insertions, deletions, substitutions))
        previous_row = current_row
    return previous_row[-1]


# [2단계-B]501개 정상 도메인 DB 파일 로드
# ⚠️ 파일명이 'correct_domain.csv'이고 'FISH_dataset' 폴더 안에 있는지 꼭 확인하세요!
csv_path = '/content/drive/MyDrive/FISH_dataset/correct_domain.csv'

print("--- 📂 정식 도메인 화이트리스트 데이터셋 로드 시작 ---")
if os.path.exists(csv_path):
    # 첫 줄 naver.com 누락 방지를 위한 header=None 처리 패치 적용
    df_domains = pd.read_csv(csv_path, header=None, names=['official_domain'])
    official_domains_db = df_domains['official_domain'].dropna().tolist()
    print(f"✅ 데이터셋 로드 완료! 총 {len(official_domains_db)}개의 정식 도메인이 준비되었습니다.")
else:
    print(f"❌ [에러] CSV 파일을 찾을 수 없습니다: {csv_path}")
    print("💡 경로가 다를 경우 아래 백업 데이터셋으로 자동 대체하여 연산합니다.")
    official_domains_db = ["naver.com", "daum.net", "google.com", "kakaobank.com", "toss.im", "coupang.com"]
```

```python
import os
import re
import easyocr
import pandas as pd
import numpy as np
from tqdm import tqdm

# ==============================================================================
# 1. Levenshtein Distance (편집 거리 알고리즘) 직접 구현
# 설명: 공식 도메인과 피싱 도메인의 철자 차이를 수학적으로 계산하여 사칭 여부를 판별한다.
# ==============================================================================
def get_levenshtein_distance(s1, s2):
    if len(s1) < len(s2):
        return get_levenshtein_distance(s2, s1)
    if len(s2) == 0:
        return len(s1)

    # 두 도메인 간의 문자 변경, 삽입, 삭제 비용을 계산하기 위한 1차원 행렬 초기화
    previous_row = range(len(s2) + 1)

    for i, c1 in enumerate(s1):
        current_row = [i + 1]
        for j, c2 in enumerate(s2):
            # 수학적 최소 편집 비용 계산 (삽입, 삭제, 교체 비용 대조)
            insertions = previous_row[j + 1] + 1
            deletions = current_row[j] + 1
            substitutions = previous_row[j] + (c1 != c2)
            current_row.append(min(insertions, deletions, substitutions))
        previous_row = current_row

    # 최종 계산된 최소 편집 거리(철자 차이 개수) 반환
    return previous_row[-1]


# ==============================================================================
# 2. 업데이트된 정식 도메인 DB 파일 로드 (화이트리스트 데이터셋)
# ==============================================================================
correct_domain_path = '/content/drive/MyDrive/FISH_dataset/correct_domain.csv'
phishing_images_dir = '/content/drive/MyDrive/FISH_dataset/AI_OpenSourceProgramming_Phishing/'
correct_images_dir = '/content/drive/MyDrive/FISH_dataset/AI_OpenSourceProgramming_Correct/'

print("--- 📂 [1/3] 업데이트된 정식 도메인 정답지 DB 로드 ---")
if os.path.exists(correct_domain_path):
    df_domains = pd.read_csv(correct_domain_path, header=None, names=['official_domain'])
    official_domains_db = df_domains['official_domain'].dropna().str.strip().str.lower().tolist()
    official_domains_db = list(set(official_domains_db))
    print(f"✅ 화이트리스트 로드 완료! (총 {len(official_domains_db)}개 도메인 대조 준비)")
else:
    print("❌ 업데이트된 correct_domain.csv 파일을 찾을 수 없습니다. 경로를 확인해 주세요.")
    official_domains_db = ["naver.com", "daum.net", "google.com"]


# ==============================================================================
# 3. 편집 거리(Levenshtein Distance) 사칭 여부 판별 함수
# ==============================================================================
def check_url_spoofing(suspect_url):
    # 🧼 [선제 차단] 메신저 ID 유도형 피싱 문자열 필터링
    lower_url = suspect_url.lower().replace(" ", "")
    if "line:" in lower_url or "line." in lower_url or "kakao:" in lower_url or "kakaotalk" in lower_url:
        return f"   🚨 경고: 메신저 ID 유도형 피싱 위험 감지! (라인/카카오톡 ID 추가 유도 패턴)"

    clean = suspect_url.lower().strip().replace(" ", "")

    # 🧼 [노이즈 필터] 진짜 인터넷 도메인 확장자가 포함되어 있는지 원본 유효성 검사
    valid_extensions = ['.com', '.kr', '.net', '.org', '.me', '.im', '.tv', '.cc', '.co']
    if not any(ext in clean for ext in valid_extensions):
        return None  # 도메인 확장자가 아예 없는 순수 문장 기호 파편은 스킵

    # 프로토콜 및 오타 처리용 기호 청소
    clean = clean.replace("https://", "").replace("http://", "").replace("httpsll", "").replace("httpll", "").replace("https:ll", "").replace("http:ll", "")

    # 메인 도메인 영역 분리 (슬래시나 EasyOCR 깨짐 기호 'll' 기준 분할)
    clean = re.split(r'[/|)|\]|>,|ll|sll]', clean)[0]

    # OCR 끝자리 왜곡 노이즈 보정 (krl -> kr)
    if clean.endswith('krl'): clean = clean[:-1]
    if clean.endswith('coml'): clean = clean[:-1]
    if clean.endswith('netl'): clean = clean[:-1]

    # 핵심 확장자 기준으로 주소 문자열 마감
    for ext in valid_extensions:
        if ext in clean:
            clean = clean.split(ext)[0] + ext
            break

    # 가짜 파편 필터링
    if len(clean) <= 3 or '.' not in clean:
        return None

    # 숙명여대 부속 기관 도메인 예외 통합 허용
    if "sookmyung" in clean:
        return f"   ✅ 안전: 숙명여자대학교 정식 공식 도메인입니다. ({clean})"

    # 🔍 [화이트리스트 대조] 엑셀 정답지 DB 파일에 완벽히 등록된 도메인인지 확인
    if clean in official_domains_db:
        return f"   ✅ 안전: 정식 등록된 공식 주소입니다. ({clean})"

    # 🔍 [알고리즘 연산] Levenshtein Distance 수학적 계산 돌리기
    is_spoofing = False
    matched_official = ""
    min_distance = 999

    for official in official_domains_db:
        # 추출 도메인과 정답 도메인 간의 '철자 차이 개수' 계산
        distance = get_levenshtein_distance(clean, official)

        # 철자가 1~2개만 교묘하게 비틀려 있는 사칭 변조 도메인 탐지 규칙
        if distance <= 2:
            is_spoofing = True
            if distance < min_distance:
                min_distance = distance
                matched_official = official

    if is_spoofing:
        # 철자 유사도 수치 계산 (수학적 시각화 지표)
        max_len = max(len(clean), len(matched_official))
        similarity = (1 - min_distance / max_len) * 100
        return f"   🚨 경고: 사칭 URL 위험 감지! 공식 기관 '{matched_official}' 주소와 {similarity:.1f}% 유사합니다. (Levenshtein Distance: {min_distance}글자 변조 포착)"
    else:
        # 정답지 엑셀 파일 화이트리스트에 없는 완벽한 제3의 미등록 외딴 피싱 도메인인 경우
        return f"   🚨 경고: 신뢰할 수 없는 출처의 미등록 피싱 의심 도메인입니다! ({clean})"


# ==============================================================================
# 4. 수집 사진 데이터셋(Phishing/Correct) 자동 전수 조사 가동
# ==============================================================================
print("\n--- 🔍 [2/3] 수집된 사진 데이터셋 스캐닝 및 사칭 검사 시작 ---")
reader = easyocr.Reader(['ko', 'en'], gpu=True)
target_folders = [('🚨 피싱 사진 폴더', phishing_images_dir), ('💚 정상 사진 폴더', correct_images_dir)]

for folder_name, folder_path in target_folders:
    print(f"\n==============================================")
    print(f"📂 {folder_name} 분석 파이프라인 가동")
    print(f"==============================================")

    if not os.path.exists(folder_path):
        print(f"❌ 폴더 경로를 찾을 수 없습니다: {folder_path}")
        continue

    image_files = [f for f in os.listdir(folder_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
    print(f"📸 발견된 이미지: 총 {len(image_files)}장")

    for img_file in image_files:
        full_img_path = os.path.join(folder_path, img_file)

        try:
            ocr_result = reader.readtext(full_img_path, detail=0)
            extracted_text = " ".join(ocr_result)

            # 주소 형태 및 메신저 키워드 복합 정규식 파싱
            urls_found = re.findall(r'(https?://[^\s]+|https?:ll[^\s]+|https?ll[^\s]+|www\.[^\s]+|[a-zA-Z0-9-]+\s?\.\s?[a-zA-Z]{2,4}[^\s]*|[lL][iI][nN][eE]\s?[\.\:]\s?[^\s]+|[kK][aA][kK][aA][oO][^\s]*)', extracted_text)

            if urls_found:
                print(f"\n📷 [파일명: {img_file}] 분석 중...")
                has_valid_url = False
                for url in list(set(urls_found)):
                    analysis_result = check_url_spoofing(url)
                    if analysis_result is not None:
                        print(f"   🔗 추출된 링크: {url}")
                        print(analysis_result)
                        has_valid_url = True
                if not has_valid_url:
                    print("   ℹ️ 이 사진 안에는 유효한 URL 링크가 없습니다.")
            else:
                pass
        except Exception as e:
            print(f"\n❌ 이미지 처리 중 오류 발생 ({img_file}): {e}")

print("\n🎉 --- [3/3] 수집한 사진 데이터셋 기반 URL 사칭 대조 실험 종료 ---")
```
